THEORY:

Basic signal operations manipulate the signal characteristics by acting on the independent
variables which are used to represent them. This means that instead of performing operations
like addition, subtraction, and multiplication between signals, we will perform them on the
independent variables. Some of the common operations are :

i) Time reversal: y[n] = -1.x[n]

ii) Time shifting: y[n] = x[n ± n0]

iii)Time scaling: y[n] = a.x[n]

iv)Time shifting plus time scaling: y[n] = a.x[n ± n0]

---



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Continuos signal
f = float(input("Enter the frequency of the signal :"))
t = np.arange (0,1,0.01)
y = np.sin(2*np.pi*f*t)
plt.plot(t,y)
plt.title("Continuos Signal-Original signal")
plt.show()

#time reversal

tr=np.zeros(len(t))
for i in range (0,len(t)):
  tr[i]=-1*t[len(t)-1-i]

yr= -1*(np.sin(2*np.pi*f*t))
plt.plot(tr,yr)
plt.show()

#shifting of signal
k = int(input("Enter the shifting factor :"))
tl = [x - k for x in t]
plt.subplot(1,2,1)
plt.plot(tl,y)
plt.title ("Left shift")
tr = [x + k for x in t]
plt.subplot(1,2,2)
plt.plot(tr,y)
plt.title ("Right shift")
plt.show()

#Scaling of signal
a = int(input("Enter the scaling factor :"))
s = [x * a for x in t]
plt.plot(s,y)
plt.title("Up-Scaling")
plt.show()

#left shifting and upscaling of signal
upl = [a*(x - k) for x in t]
plt.subplot(1,2,1)
plt.plot (upl,y)
plt.title("Left shifting,then Up-Scaling")

#right shifting and upscaling of signal
upr = [a*(x + k) for x in t]
plt.subplot(1,2,2)
plt.plot (upr,y)
plt.title("Right shifting,then Up-Scaling")
plt.show()
```

```python
#Discrete signal
u,l,step = input("Enter the input sequence:").split(",",3)
t=np.arange(float(u),float(l)+1,float(step))
y=np.zeros(len(t))
for i in range (0,len(t)):
  y[i]=t[i]
plt.stem(t,y)
plt.title("Discrete Signal-Original signal")
plt.show()

#time reversal
tr=np.zeros(len(t))
for i in range (0,len(t)):
  tr[i]=-1*t[len(t)-1-i]
yr=np.zeros(len(t))
for i in range (0,len(t)):
  yr[i]=1*t[len(tr)-1-i]
plt.stem(tr,yr)
plt.show()

#shifting of signal
k = int(input("Enter the shifting factor :"))
tl = [x - k for x in t]
plt.subplot(1,2,1)
plt.stem(tl,y)
plt.title ("Left shift")
tr = [x + k for x in t]
plt.subplot(1,2,2)
plt.stem(tr,y)
plt.title ("Right shift")
plt.show()


#Scaling of signal
a = int(input("Enter the scaling factor :"))
s = [x * a for x in t]
plt.stem(s,y)
plt.title("Up-Scaling")
plt.show()

#left shifting and upscaling of signal
upl = [a*(x - k) for x in t]
plt.subplot(1,2,1)
plt.stem (upl,y)
plt.title("Left shifting,then Up-Scaling")

#right shifting and upscaling of signal
upr = [a*(x + k) for x in t]
plt.subplot(1,2,2)
plt.stem (upr,y)
plt.title("Right shifting,then Up-Scaling")
plt.show()

```
